```python
df = df_train.copy()
```

```python
#Split the train data into train and validation 

from sklearn.model_selection import train_test_split

X = df["message"]
y = df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
X_train = df_train['message']
X_test = df_test['message']
y_train = df_train['sentiment']
```

```python
text_clf_1 = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LinearSVC(penalty='l2', loss='squared_hinge', dual=True, tol=0.001, C=0.8, multi_class='ovr', fit_intercept=True, intercept_scaling=1, class_weight=None, verbose=0, random_state=None, max_iter=1000))
])

# Feed the training data through the pipeline
text_clf_1.fit(X_train, y_train)
```

```python
# Form a prediction set
predictions = text_clf_1.predict(X_test)
```

```python
# Report the confusion matrix
from sklearn import metrics
print(metrics.confusion_matrix(y_test,predictions))
```

```python
# Print a classification report
print(metrics.classification_report(y_test,predictions))
```

```python
# Print the overall accuracy
print(metrics.accuracy_score(y_test,predictions))
```

```python
text_clf_3 = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', SVC()),
])

# Feed the training data through the pipeline
text_clf_3.fit(X_train, y_train)
```

```python
kernels = ['Polynomial', 'RBF', 'Sigmoid','Linear']#A function which returns the corresponding SVC model
def getClassifier(ktype):
    if ktype == 0:
        # Polynomial kernal
        return SVC(kernel='poly', degree=8, gamma="auto")
    elif ktype == 1:
        # Radial Basis Function kernal
        return SVC(kernel='rbf', gamma="auto")
    elif ktype == 2:
        # Sigmoid kernal
        return SVC(kernel='sigmoid', gamma="auto")
    elif ktype == 3:
        # Linear kernal
        return SVC(kernel='linear', gamma="auto")
```
